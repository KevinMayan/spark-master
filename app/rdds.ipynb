## Importando Utilitários

```python
import findspark
findspark.init()
from pyspark.sql import SparkSession
```

## Criar Sessão Spark

```python
spark = SparkSession.builder.appName("RDD").getOrCreate()
```

## Criando um Resilient Distributed Dataset (RDD)

```python
data = [1, 2, 3, 4, 5]

# 2 is the number of partitions to cut the dataset
rdd = spark.sparkcontext.parallelize(data, 2)
```

## Aplicando a Função Map no RDD

```python
rdd2 = rdd.map(lambda x: x * 10)

rdd2.collect()
```

## Criando RDD com Arquivo de Texto

```python
filename = "data/partitions.txt"

rdd3 = spark.sparkcontext.textFile(filename)

rdd3.map(lambda x: len(x)).reduce(lambda a, b: a + b)
```

## Terminando Sessão Spark

```python
spark.stop()
```
